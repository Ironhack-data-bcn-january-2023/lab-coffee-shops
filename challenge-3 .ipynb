# 2. Plot a map with the coffee shops

```python
import geopandas as gdp
from cartoframes.viz import Map, Layer, popup_element
```

```python

import folium
from folium import Choropleth, Circle, Marker, Icon, Map
from folium.plugins import HeatMap, MarkerCluster
import pandas as pd
```

```python
bcn_lat = 41.3961786
bcn_lon = 2.1914234

bcn_map = Map(location=[bcn_lat, bcn_lon], zoom_start = 15)
bcn_map
```

```python
# Convert the dataframe to a GeoDataFrame
gdf = gdp.GeoDataFrame(df_coffe, geometry=gdp.points_from_xy(df_coffe.lon, df_coffe.lat))
# Create a folium map centered in Barcelona
bcn_map = folium.Map(location=[bcn_lat, bcn_lon], zoom_start=15)
# Loop through each row of the GeoDataFrame and add a marker for each coffee shop
for idx, row in gdf.iterrows():
    folium.Marker(location=[row.lat, row.lon], popup=row.name).add_to(bcn_map)
# Display the map
bcn_map
```
